```python
plt.plot([1,2,3,4,5],[1,4,9,16,25],'ro')
plt.title('Squares of the numbers')
#  labels
plt.xlabel('The Values')
plt.ylabel('The Values')

plt.show()
```

```python
plt.plot([1,2,3,4,5],[1,4,9,16,25],'ro-')
plt.title('Squares of the numbers')
#  labels with colors
plt.xlabel('The Values',fontsize=18,color='red')
plt.ylabel('The Values')

plt.show()
```

```python
plt.plot([1,2,3,4,5],[1,4,9,16,25],'ro-')
plt.title('Squares of the numbers')
# cutomise the start and end of axis
#plt.axis(['startofx','endofx','startofy','endofy'])
plt.axis([0,6,0,45])
plt.xlabel('The Values')
plt.ylabel('The Values')

plt.show()
```

```python
#increse the thickness of line
plt.plot([1,2,3,4,5],[1,4,9,16,25],'ro-',linewidth=3.0)
plt.title('Squares of the numbers')

plt.axis([0,6,0,45])
plt.xlabel('The Values')
plt.ylabel('The Values')

plt.show()
```

```python
#show grid
plt.plot([1,2,3,4,5],[1,4,9,16,25],'ro-')
plt.title('Squares of the numbers')
plt.grid(True)

plt.show()
```

```python
#show multiple plots
x=[i for i in range(20)]
#error, can't do it in list so we create a numpy array

import numpy as ny
x=ny.array(x)
plt.plot(x,x**3,'r-',x,x**2,'b-',x,x**4,'y-')
plt.show()
```

```python
#create a list with random values
import random
x=[random.randint(1,100) for i in range(100)]
y=[random.randint(1,100) for i in range(100)]
plt.scatter(x,y)
plt.axis([0,100,0,100])
plt.show()
```

```python
#bar
import random
x=[random.randint(1,100) for i in range(100)]
y=[random.randint(1,100) for i in range(100)]
plt.bar(x,y)
plt.axis([0,100,0,100])
plt.show()
```
